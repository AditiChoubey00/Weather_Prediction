```python
#Ridge regression penalises the coefficients to have colinearity
#has a parameter called lambda but it is not used since lamda is a keyword in python
#Alpha here states how much the coefficients here are penalized
from sklearn.linear_model import Ridge
rr = Ridge(alpha=.1)
```

```python
# This gives all the columns except the given three in square brackets
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
```

```python
#here weather is the weather,ridge regression is model,predicteed is all col except name,station n target
#start states that we need atleast 3650 amount of data before starting the prediction
#step states that prediction will be made after each 90 days
def backtest(weather, model, predictors, start=3650, step=90):
  #this list will have dataframes of next 90 days
    all_predictions = []
      #start=3650,weather.shape[0] is the end of the weather, advance with step =90
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]
        model.fit(train[predictors], train["target"])
        #preds will return the result as numpy array
        preds = model.predict(test[predictors])
        #to make it easy to work with we have converted into pandas.series
        preds = pd.Series(preds, index=test.index)
        #here we concatenate the actual test result and the predicted test results and give axis=1 i.e. treat both of
        #as separate columns
        combined = pd.concat([test["target"], preds], axis=1)
        #create the columns that will be formed after concatenation
        #create a column called diff that will be the diff between actual and predicted values and will convert it into absolute values
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
predictions.sort_values("diff", ascending=False)
```

```python
def pct_diff(old, new):
    return (new - old) / old

def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean)
```

```python
weather = weather.iloc[14:,:]
weather = weather.fillna(0)
```

```python
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
```

```python
predictions = backtest(weather, rr, predictors)
mean_absolute_error(predictions["actual"], predictions["prediction"])
```
